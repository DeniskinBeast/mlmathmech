# Введение в машинное обучение

## Введение

В первых двух частях мы попрактиковались в первичном анализе данных с `Pandas` и в построении картинок, позволяющих делать выводы по данным. Сегодня наконец перейдем к машинному обучению. 

Наверно хочется сразу рвануть в бой, но сначала поговорим про то, какую именно задачу будем решать и каково ее место в области машинного обучения.
Классическое, общее определение машинного обучения звучит так (T. Mitchell "Machine learning", 1997): "Говорят, что компьютерная программа *обучается* при решении какой-то задачи из класса *T*, если ее производительность, согласно метрике *P*, улучшается  при накоплении опыта *E*".

T - Задача (Task) <br> P - Mетрика (Performance) <br> E - Oпыт (Experience)

Далее в разных сценариях под *T, P*,  и *E* подразумеваются совершенно разные вещи. Среди самых популярных задач *T* в машинном обучении:
 - __классификация__ – отнесение объекта к одной из категорий на основании его признаков 
 - __регрессия__ – прогнозирование количественного признака объекта на основании прочих его признаков 
 - __кластеризация__ – разбиение множества объектов на группы на основании признаков этих объектов так, чтобы внутри групп объекты были похожи между собой, а вне одной группы  – менее похожи
 - __детекция аномалий__ – поиск объектов, "сильно непохожих" на все остальные в выборке либо на какую-то группу объектов
 - и много других, более специфичных. Хороший обзор дан в 5 главе книги "Deep Learning" (Ian Goodfellow, Yoshua Bengio, Aaron Courville, 2016) 
 
Под опытом *E* понимаются данные (без них никуда), и в зависимости от этого алгоритмы машинного обучения могут быть поделены на тех, что __обучаются с учителем__ и __без учителя__ (supervised & unsupervised learning). В задачах обучения без учителя имеется __выборка__, состоящая из __объектов__,  описываемых набором __признаков__.  В задачах обучения с учителем вдобавок к этому для каждого объекта некоторой выборки, называемой __обучающей__, известен __целевой признак__ – по сути это то, что хотелось бы прогнозировать для прочих объектов, не из обучающей выборки.



### Формальная постановка задачи машинного обучения

Основные понятия и обозначения

$X$ - множество объектов <br>
$Y$ - множество ответов <br>
$y: X \rightarrow Y$ - неизвесная зависимость (target function)

__Дано:__ 

$ \{  x_1, ..., x_l \} \subset X$ - обучающая выборка (training samples);

$ y_{i} = y(x_i), i = 1 ... l $ - известные ответы

__Найти:__

$a: X \rightarrow Y$ - алгоритм, решающую функцию (decision function), приближающую $y$ на всем множестве X.

Как задаются объекты? Признаковое описание. 

$f_j: X \rightarrow D_j, j = 1, ..., n$ - признаки объектов (features)
Типы признаков: 
- `Бинарные`: $\{0, 1\}$
- `Категориальные`: *{ Пол, Город, Страна, Персонаж в Dota2 и т.д. }*
- `Порядковые`: *{ Уровень образования(Бакалавр, Магистр,...), Ранг в CSGO и т.д. }*
- `Количественные`: $f(x) \in \mathbb{R}$

Вектор $(f_1(x),..., f_n(x))$ - признаковое описание объекта. 

Матрица "объекты-признаки" и называется, собственно говоря, данными в задачах анализа данных. 




### Задача обучения с учителем (Supervised Learning)
Задачи обучения с учителем могут разделяться на виды, по типу целевого признака.
Перечислим основные из них.

__Задачи классификации__ (classification):

- $Y = \{0,1\}$ - задача бинарной классификации. 

- $Y = \{1,...,M\}$ - многоклассовая (multi-class) классификация. Например: определить предметную область для научной статьи (математика, биология, психология и т.д.) 

- $Y = \{0,1\}^{M}$ - многоклассовая классификация с пересекающимися классами (multi-label classification). Например: классифицировать текст набором тегов (#онлайн_курс, #машинное_обучение, #классификация, ...)

__Задачи восстановления регрессии__ (regression):

- $Y \in \mathbb{R}$ или $R^{m}$ - определить стоимость дома. 

__Задачи ранжирования__ (ranking):

- $Y$ - упорядоченное множество. Задача ранжирования поисковой выдачи, отсортировать документы по релевантности. 

### Задача обучения без учителя (Unsupervised Learning) 
Задачи обучения без учителя - это класс задач, где ответы неизвестны или вообще не существуют, и требуется найти некоторые закономерности в данных лишь на основе признаковых описаний: 

- __Кластеризация__ - задача разделения объектов на группы, обладающие некоторыми свойствами. Пример: кластеризация абонентов мобильного оператора. 

- __Оценивание плотности__ - приближение распределения объектов. Например: задача обнаружения аномалий, в которой на этапе обучения известны лишь примеры "правильного" поведения, а в дальнейшем требуется обнаружить случаи некорректного поведения. 

- __Визуализация__ - задача изображения многомерных объектов в двухмерном или трехмерном пространстве таким образом, чтобы сохранялось как можно больше зависимостей и отношений между ними. 

- __Понижение размерности__ - задача генерации новых признаков, таких что их меньше, чем исходных, но при этом с их помощью задача решается не хуже (или с небольшими потерями качества, или лучше - зависит от подстановки). Это задачи тематического моделирования и построения рекомендаций.

## Задача классификации

#### Генерация данных

```python
np.random.seed(1337)
n = 100
a = np.random.normal(loc=0, scale=1, size=(n, 2))
b = np.random.normal(loc=3, scale=2, size=(n, 2))
X = np.vstack([a, b])
y = np.hstack([np.zeros(n), np.ones(n)])
print(X.shape, y.shape)
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=50, alpha=1)
plt.xlabel('x1')
plt.xlabel('x2')
plt.grid()
```

## k-Nearest Neighbors (kNN)

Метод ближайших соседей (k Nearest Neighbors, или kNN) — тоже очень популярный метод классификации, также иногда используемый в задачах регрессии. Это, наравне с деревом решений, один из самых понятных подходов к классификации. На уровне интуиции суть метода такова: посмотри на соседей, какие преобладают, таков и ты. Формально основой метода является гипотеза компактности: если метрика расстояния между примерами введена достаточно удачно, то схожие примеры гораздо чаще лежат в одном классе, чем в разных.

[KNeighborsClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier) - Классификатор работающий методом ближайших соседей.   

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
clf_knn = KNN(n_neighbors=5) # получаем экземпляр модели
clf_knn.fit(X, y) # обучаем модель на данных
```

```python
# Предсказываем значения класса для точек на плоскости
clf_knn.predict([[2, 0],
                 [10, 20],
                 [-10, -20]])
```

```python
# Предсказываем значение, и получаем вероятность приндлажености классам
clf_knn.predict_proba([[2, 0],
                       [0, 2],
                       [-10, -20]])
```

```python
def get_grid(data, border=1., step=.01):
    x_min, x_max = data[:, 0].min() - border, data[:, 0].max() + border
    y_min, y_max = data[:, 1].min() - border, data[:, 1].max() + border
    return np.meshgrid(np.arange(x_min, x_max, step),
                       np.arange(y_min, y_max, step))

def plot_model(X, y, clf, proba=False):
    clf.fit(X, y) # обучаем модель
    xx, yy = get_grid(X)
    if proba: # предсказываем значения для каждой точки плоскости 
        predicted = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1].reshape(xx.shape)
    else:
        predicted = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    plt.figure(figsize=(8, 8))
    plt.pcolormesh(xx, yy, predicted, cmap='spring')
    plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap='spring', edgecolors='b')
    plt.ylim([yy.min(),yy.max()])
    plt.xlim([xx.min(),xx.max()])
    return clf
```

```python
len(get_grid(X))
```

```python
%%time
plot_model(X, y, clf_knn, proba=True)
```

## Decision Tree

# Unsupervised Learning
## K-mean

```python
# Начнём с того, что насыпем на плоскость три кластера точек
X_un = np.zeros((150, 2))

np.random.seed(seed=42)
X_un[:50, 0] = np.random.normal(loc=0.0, scale=.3, size=50)
X_un[:50, 1] = np.random.normal(loc=0.0, scale=.3, size=50)

X_un[50:100, 0] = np.random.normal(loc=2.0, scale=.5, size=50)
X_un[50:100, 1] = np.random.normal(loc=-1.0, scale=.2, size=50)

X_un[100:150, 0] = np.random.normal(loc=-1.0, scale=.2, size=50)
X_un[100:150, 1] = np.random.normal(loc=2.0, scale=.5, size=50)

plt.figure(figsize=(5, 5))
plt.plot(X_un[:, 0], X_un[:, 1], 'bo');
```

```python
from sklearn.cluster import KMeans
plot_model(X_un, [], KMeans(n_clusters=3))
```

### Unsupervised k-NN

```python
from sklearn.neighbors import NearestNeighbors
samples = [[0, 0, 2], [1, 0, 0], [0, 0, 1]]

neigh = NearestNeighbors(2, 0.4)
neigh.fit(samples) 

neigh.kneighbors([[0, 0, 1.3]], 2, return_distance=False)
```
